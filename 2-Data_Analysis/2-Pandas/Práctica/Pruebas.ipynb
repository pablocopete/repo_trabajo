```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('planets')
df
```

```python
# Vamos a coger el periodo orbital, solo la parte enter y dejar a true si es par o a False si es impar
```

```python
PO_PAR = []
for index, row in df.iterrows():
    if pd.isna(row['orbital_period']):
        print('hola')
        PO_PAR.append(np.nan)
        continue
    
    numerito_int = int(row['orbital_period']) 
    if numerito_int %2 == 0:
        PO_PAR.append(True)
    else:
        PO_PAR.append(False)

PO_PAR
```

```python
def popar(orbita):
    if pd.isna(orbita):
        return np.nan
    if int(orbita) %2 == 0:
        return True
    else:
        return False
```

```python
df.loc[:, 'orbital_period'].apply(popar)
```

```python
def abreviatura(method):
    palabras = method.split()
    abrv = ''
    for palabra in palabras:
        abrv += palabra[0]
    return abrv

abreviatura('Hola Pablo')

df.loc[:,'method'].apply(abreviatura)
```

```python
def encuentra_zipcode(direccion):
    for index, char in enumerate(direccion):
        if char.isdigit():
            if direccion[index:index + 5].isdigit():
                return direccion[index:index + 5]

```

```python
def geolocaliza_zipcode(zip_code):
    geolocalizador = geopy.Nominatim(user_agent='Tutorial')
    respuesta = geolocalizador.geocode(f'{zip_code}, Spain')
    return respuesta.raw['lat'], respuesta.raw['lon']

def geo_address(address):
    return geolocaliza_zipcode(encuentra_zipcode(address))
```
