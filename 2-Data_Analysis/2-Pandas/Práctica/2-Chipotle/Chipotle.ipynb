### Step 5. What is the price of each item? Get only item_name and unit_price

```python
df
```

```python
# Crea una nueva columna llamada unit_price
df2 = df.groupby('item_name')['item_price'].mean().to_frame()
df2 = df2.reset_index()
df3 = pd.merge(left=df, right=df2 ,left_on='item_name', right_on='item_name', how='inner')
```

```python
print(df3.columns)
df3.columns = ['order_id', 'quantity', 'item_name', 'choice_description', 'item_price', 'mean_item_price']
df3
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
df3.loc[:,'item_price'].max()
```

```python
df3.sort_values('item_price', ascending=False)
```

```python
mask = df3.loc[:,'item_name'] == df3.sort_values('item_price', ascending=False).iloc[0,:]['item_name']
df3.loc[mask, :]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
mask = df3.loc[:,'item_name'] == 'Veggie Salad Bowl'
print(mask.sum())
df3.loc[mask, :].count()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
mask = (df3.loc[:, 'item_name'] == 'Canned Soda') & (df3.loc[:, 'quantity'] > 1)
df3.loc[mask, :].count().max()
```
