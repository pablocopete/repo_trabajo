```python
import undetected_chromedriver as uc

options = uc.ChromeOptions()
options.add_argument("--no-sandbox")

driver = uc.Chrome(options=options)
```

```python
url = 'https://www.filmaffinity.com/es/topgen.php'
```

```python
# Meterse en la web
driver.get(url)

# Rechazar cookies
boton_rechazo_html = '//*[@id="disagree-btn"]'
boton_rechazo = driver.find_element(By.XPATH, boton_rechazo_html)
boton_rechazo.click()
```

```python
n_reps = 250//30 + 1
```

```python
for _ in range(n_reps):
    # Bajamos hasta abajo de la pagina
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    # Cogemos el boton y pulsamos
    boton_html = '//*[@id="load-more-bt"]/i'
    boton = driver.find_element(By.XPATH, boton_html)
    boton.click()
    # Esperamos
    time.sleep(2)
```

```python
soup = bs(driver.page_source,"lxml")
```

```python
soup
```

```python
titulo = []
year = []
director = []
actores = []
nota = []

for i in range (5):
    todo = soup.select('div.mc-info-container')[i].get_text()
    print(todo)
    number_to_delete = todo.count('')
    todo = todo.split('\n')
    my_list = []
    for j in range(len(todo)):
        if todo[j] != '':
            my_list.append(todo[j])
    
    print(len(my_list))
    print(my_list)

    if len(my_list) == 4:  
        titulo.append(my_list[0])
        year.append(my_list[1])
        director.append(my_list[2])
        actores.append(my_list[3])
    elif len(my_list) == 5:  
        titulo.append(my_list[0])
        year.append(my_list[2])
        director.append(my_list[3])
        actores.append(my_list[4])
    else:
        print('Resultado inesperado')

print(f'Titulo: {titulo}')
print(year)
print(actores)
print(director)
```
