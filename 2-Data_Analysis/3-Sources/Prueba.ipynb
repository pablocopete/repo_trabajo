```python
import pandas as pd
import numpy as np
import json
import requests
```

```python
url = 'https://valencia.opendatasoft.com/api/explore/v2.1/catalog/datasets/centros-educativos-en-valencia/records?limit=100'
response = requests.get(url)
response.text
```

```python
my_dicts = json.loads(response.text)
my_dicts['results']
```

```python
df_feo = pd.DataFrame(my_dicts['results'])
df_feo.head()
# df_feo.iloc[0,0]
```

```python
df_bonito = df_feo.loc[:, ['dlibre', 'regimen', 'codpos', 'direccion', 'geo_point_2d']]
df_bonito.columns = ['nombre', 'tipo', 'CP', 'direccion', 'coordenadas']
df_bonito
```

```python
df_bonito.info()
```

```python
df_bonito.groupby(['CP', 'tipo'])['nombre'].count()
```

```python
max_lat = -10000000000
max_lat_nombre = ''
min_lat = +1000000000
min_lat_nombre = ''
max_lon = -10000000000
max_lon_nombre = ''
min_lon = +1000000000
min_lon_nombre = ''

for index, row in df_bonito.iterrows():
    if row['coordenadas']['lat'] > max_lat:
        max_lat = row['coordenadas']['lat']
        max_lat_nombre = row['nombre']

    if row['coordenadas']['lat'] < min_lat:
        min_lat = row['coordenadas']['lat']
        min_lat_nombre = row['nombre']

    if row['coordenadas']['lon'] > max_lon:
        max_lon = row['coordenadas']['lon']
        max_lon_nombre = row['nombre']
    
    if row['coordenadas']['lon'] < min_lon:
        min_lon = row['coordenadas']['lon']
        min_lon_nombre = row['nombre']
    

print(max_lat_nombre, min_lat_nombre, max_lon_nombre, min_lon_nombre)
```

## Forma alternativa de montar la tabla

```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
url = 'https://valencia.opendatasoft.com/api/explore/v2.1/catalog/datasets/centros-educativos-en-valencia/records?limit=100'
response = requests.get(url)
```

```python
data =json.loads(response.text)['results']
data
```

```python
lat = []
lon = []
nombre = []
CP = []
tipo = []
direccion = []

for dictionary in data:
    lon.append(dictionary['geo_point_2d']['lon'])
    lat.append(dictionary['geo_point_2d']['lat'])
    nombre.append(dictionary['dlibre'])
    CP.append(dictionary['codpos'])
    tipo.append(dictionary['regimen'])
    direccion.append(dictionary['direccion'])


dict_df = {
    'nombre': nombre,
    'tipo': tipo,
    'CP': CP,
    'direccion': direccion,
    'lon' : lon,
    'lat': lat
}

df = pd.DataFrame(dict_df)
df
```

```python
mask1 = df.loc[:,'lon'] == df.loc[:,'lon'].max()
df.loc[mask1,['nombre', 'lon']]
```
