# Ataca la API de RICK Y MORTY

Consultar la [documentaci√≥n](https://rickandmortyapi.com/documentation)

Quiero que os traigais y pongais en un dataframe (valora que campos quieres mantener y que campos no!) todos los **personajes** de la serie que sean:

- De especie humana
- Que esten de estatus muerto
- Y de genero female

```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
# Hay un comando para traer todos de golpe, pero prefiero traer por batches para practicar
lista_jsons = []
n_pages = (826 // 20) +1

for i in range (1, n_pages+1):
    url = f'https://rickandmortyapi.com/api/character/?page={i}'
    response = requests.get(url)
    lista_jsons.append(json.loads(response.text))
```

```python
lista_jsons[0]
```

```python
lista_df = []

for page in lista_jsons:
    name = []
    status = []
    species = []
    gender = []
    for character in page['results']:
        name.append(character['name'])
        status.append(character['status'])
        species.append(character['species'])
        gender.append(character['gender'])

    df = pd.DataFrame({
        'name': name,
        'status': status,
        'species': species,
        'gender': gender
        })
    
    lista_df.append(df)

lista_df[0]
```

```python
# Probamos con concat
df = pd.concat(lista_df, ignore_index=True)
df
```

```python
df.info()
```

```python
df.loc[:,'status'].unique()
```

```python
'''
- De especie humana
- Que esten de estatus muerto
- Y de genero female
'''

mask1 = df.loc[:, 'species'] == 'Human'
mask2 = df.loc[:,'status'] == 'Dead'
mask3 = df.loc[:, 'gender'] == 'Female'

mask = mask1 & mask2 & mask3

df.loc[mask, :]
```

```python
df.loc[mask,'name'].nunique()
```

```python
print(f'n_repetidos: {df.loc[mask,:].shape[0] - df.loc[mask,"name"].nunique()}')
```

```python
url = f"https://pokeapi.co/api/v2/pokemon/?offset={i*20}&limit=20"
response = requests.get(url)
data = json.loads(response.text)

n_pokemon = data['count']
n_pages = 1302 // 20 + 1
pokemon_nombre = []
pokemon_types = []
```

```python
for i in range(n_pages+1):
    url = f"https://pokeapi.co/api/v2/pokemon/?offset={i*20}&limit=20"
    response = requests.get(url)
    data = json.loads(response.text)['results']
    for pokemon in data:
        pokemon_nombre.append(pokemon['name'])
        url_pokemon = pokemon['url']
        pokemon_types_list_dicts = json.loads(requests.get(url_pokemon).text)['types']
        pokemon_types_list = []
        for slot in pokemon_types_list_dicts:
            pokemon_types_list.append(slot['type']['name'])
        
        pokemon_types.append(pokemon_types_list)
        
```

```python
len(pokemon_types[0])
```

```python
len(pokemon_nombre)
# pokemon_types
tipo1 = []
tipo2 = []

for index, type_couple in enumerate(pokemon_types):
    if len(type_couple) == 1:
        tipo1.append(type_couple[0])
        tipo2.append(None)
    elif len(type_couple) == 2:
        tipo1.append(type_couple[0])
        tipo2.append(type_couple[1])
    else:
        print('Ha pasado algo')
        print(index, len(type_couple))
```

```python
print(len(pokemon_nombre), len(tipo1), len(tipo2))

df =pd.DataFrame({
    'nombre': pokemon_nombre,
    'tipo 1': tipo1,
    'tipo 2': tipo2
    })

df.head()
```
